```python
from matplotlib.pyplot import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
plt.rcParams['figure.figsize'] = (9, 6)
A = imread(os.path.join("../Images/dog1.jpg"))
X = np.mean(A, -1)
img = plt.imshow(X)
img.set_cmap("gray")
plt.axis('off')
plt.show()
print(X.shape)
```

```python
from sklearn.decomposition import PCA
r = 50
pca = PCA(r)
Xapprox = pca.fit_transform(X)
print(Xapprox.shape)
```

```python
img = plt.imshow(pca.inverse_transform(Xapprox))
img.set_cmap("gray")
plt.axis("off")
plt.show()
```

## Giảm chiều dữ liệu

```python
for r in (20, 50, 100, 150, 200):
    pca = PCA(r)
    Xapprox = pca.fit_transform(X)
    
    plt.title("r =" + str(pca.n_components_))
    print(Xapprox.shape)
    img = plt.imshow(pca.inverse_transform(Xapprox))
    img.set_cmap("gray")
    plt.axis("off")
    plt.show()
    
    
    
```
