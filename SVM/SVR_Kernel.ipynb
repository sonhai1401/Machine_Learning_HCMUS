```python
y = np.array([0.96101284, 0.2415537, 0.25846608, 0.36762907, 0.48615592,
              -0.6262452, 0.64025741, 0.65477594, 0.8324045, 0.92479558,
              2.02978015, 0.99651699, 0.99683044, 0.99995752, 0.97941008,
              0.00497547, 0.96920624, 0.92358315, 0.91018836, 0.90276912,
              2.14437202, 0.59680595, 0.51094802, 0.38943151, 0.27458801,
              0.86368234, -0.26469059, -0.6272056, -0.36471195, -0.41187662,
              -1.8368655, -0.75770071, -0.83174207, -0.90513324, -0.9591322,
              -1.55498376, -0.9932663, -0.9879695, -0.9987965, -0.96345159])
```

```python
svr_rbf = SVR(kernel="rbf", C=100, gamma=0.1)
svr_lin = SVR(kernel="linear", C=100)
svr_poly = SVR(kernel="poly", C=100, degree=3, gamma="auto", coef0=1)
```

```python
svr_rbf.fit(X, y)
yhat_rbf = svr_rbf.predict(X)
```

```python
svr_lin.fit(X, y)
yhat_lin = svr_lin.predict(X)
```

```python
svr_poly.fit(X, y)
yhat_poly = svr_poly.predict(X)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.plot(X ,yhat_rbf, color="r")
plt.scatter(X,y)
plt.title("RBF")

plt.subplot(1,3,2)
plt.plot(X ,yhat_lin, color="g")
plt.scatter(X,y)
plt.title("Linear")

plt.subplot(1,3,3)
plt.plot(X ,yhat_poly, color="b")
plt.scatter(X,y)
plt.title("Poly")
```
